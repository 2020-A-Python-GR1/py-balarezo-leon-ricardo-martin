```python
#Imports

import json
import glob
import os
from os import system
```

```python
#CREAR MARCA
def crear_marca():
    input_create_marca = input("ingrese una nueva marca:")
    if os.path.exists(f"./{input_create_marca}.json"):
        print("El archivo ya existe")
    else:
        pais = input("Pais de origen: ")
        anio = input("Año de fundacion: ")
        fundador = input("Fundador: ")

        #json_file = []

        data = {}

        data = {"pais": pais,
                "anio_fundacion":anio,
                "fundador":fundador,
                "celulares" : []
        }


        with open(f'{input_create_marca}.json', 'w') as file:
            json.dump(data, file, indent=3)
        
```

```python
#LEER CELULARES
def def_leer_celular():
    listar_marcas()
    leer_marca = input("¿De qué Marca desea mostrar información de sus celulares?\n")
    if(os.path.exists(f'./{leer_marca}.json')):
        try:
            with open(f'./{leer_marca}.json') as file:
                data = json.load(file)
                #print(data[])

                for celular in data['celulares']:
                    print('modelo:', celular['modelo'])
                    print('gama:', celular['gama'])
                    print('anio:', celular['anio'])
                    print('precio:', celular['precio'])
                    print("\n")
        except Exception as Error:
            print("error leyendo el archivo")
    else:
        print("La Marca no EXISTE")
```

```python
#LEER MARCA
def def_leer_marca():
    listar_marcas()
    leer_marca = input("¿De qué Marca desea mostrar información?\n")
    if(os.path.exists(f'./{leer_marca}.json')):
        try:
            with open(f'./{leer_marca}.json') as file:
                data = json.load(file)
                #print(data[])

                print("Pais: ",data['pais'])
                print("Año de Fundacion: ",data['anio_fundacion'])
                print("Fundador: ",data['fundador'])

        except Exception as Error:
            print("error leyendo el archivo")
    else:
        print("La Marca no EXISTE")
```

```python
#Actualizar MARCA
def def_actualizar_marca():
    listar_marcas()
    leer_marca = input("¿De qué Marca desea actualizar la información?\n")
    if(os.path.exists(f'./{leer_marca}.json')):
        
        with open(f'{leer_marca}.json') as json_file: 
            data = json.load(json_file)
            
            nuevo_pais = input("Pais: ")
            data['pais'] = nuevo_pais
            
            nuevo_anio = input("Año de Fundacion: ")
            data['anio_fundacion'] = nuevo_pais
            
            nuevo_fundador = input("Fundador: ")
            
            data['fundador'] = nuevo_fundador
        
        with open(f'{leer_marca}.json', 'w') as json_file:
            json_file.write(json.dumps(data, indent=3))
            
            #json.dump(data, file, indent=3)
            
       

        
    else:
        print("La Marca no EXISTE")
        
#def_actualizar_marca()
```

```python
#Borrar MARCA

#crear_marca()
def borrar_marca():
    listar_marcas()
    input_delete_celular = input("¿Que Marca desea borrar: ?")
    if os.path.exists(f"./{input_delete_celular}.json"):
        try:
            os.remove(f"./{input_delete_celular}.json")
            print("Marca eliminada!")
        except Exception as Error:
            print("Error Elimiando el archivo")
    else: 
        print("No existe el la Marca !!!")
        

```

```python
#Listar MARCAS

def listar_marcas():
    jsfiles = []
    for file in glob.glob("./*.json"):
        
        marca = file[2:-5]
        #print(marca,"\n")
        jsfiles.append(marca)

    print("Las Marcas existentes son: ",jsfiles) 
    #print(type(jsfiles))

#listar_marcas()
```

```python
#actualizar CELULAR

def def_actualizar_celular():
    listar_marcas()
    
    input_celular = input("¿De qué Marca desea actualizar la información del celular?\n")
    
    if os.path.exists(f'./{input_celular}.json'):

        with open(f'./{input_celular}.json') as file:
            data = json.load(file)

        print("Los Modelos disponibles para la Marca seleccionado son: ")
        i = 0
        for celular in data['celulares']:

            print('(', i, ')','modelo:', celular['modelo'])
            i = i + 1

        try:
            indice = int(input("Seleccione el Indice del modelo que desea actualizar: "))

            with open(f'./{input_celular}.json') as json_file: 
                data = json.load(json_file)

                nuevo_modelo = input("Modelo: ")
                data['celulares'][indice]['modelo'] = nuevo_modelo

                nuevo_gama = input("Gama: ")
                data['celulares'][indice]['gama'] = nuevo_gama

                nuevo_anio = input("Año: ")      
                data['celulares'][indice]['anio'] = nuevo_anio

                nuevo_precio = input("Precio: ")      
                data['celulares'][indice]['precio'] = nuevo_precio


            with open(f'./{input_celular}.json', 'w') as json_file:
                json_file.write(json.dumps(data, indent=3))

        except:
            print("INGRESE SOLO VALORES NUMERICOS")
            def_actualizar_celular()
    
    else:
        print("No existe la marca")
    
    
    
#def_actualizar_celular()   
    
```

```python
def menu():
    print("DEBER DE OPERACIONES CRUD - BALAREZO MARTIN")
    print("Seleccione la opcion que desea realizar")
    print("1. Crear Marca")
    print("2. Mostrar Marcas")
    print("3. Actualizar Marca")
    print("4. Eliminar Marca")
    
    print("5. Crear Celular")
    print("6. Mostrar Celulares")
    print("7. Actualizar Celular")
    #print("4. Eliminar Celular")
    
    
    opcion = input("Opcion N°: ")
    if(opcion == "1" or opcion == "2" or opcion == "3" or opcion == "4" or opcion == "5" or opcion == "6" or opcion == "7"):
        CRUD(opcion)
        
    else:
        print("Opcion equivocada")
        menu()
```

```python
def CRUD(seleccion = "1"):
    def devolver_seleccion():
        
        opciones = {
            "1": crear_marca,
            "2": def_leer_marca,
            "3": def_actualizar_marca,
            "4": borrar_marca,
            "5": crear_celular,
            "6": def_leer_celular,
            "7": def_actualizar_celular
        }
        
        return opciones[seleccion]()
    
    return devolver_seleccion()
```
